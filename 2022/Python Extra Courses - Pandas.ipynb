Argument penting lain dalam drop_duplicates() adalah <code>keep</code> , yang bisa memiliki nilai:

    first: (default) Drop duplikat kecuali temuan pertama

```python
temp_df2 = movies_df.append(movies_df)  # make a new copy

temp_df2.drop_duplicates(inplace=True, keep=False)

temp_df2.shape
```

### Dropna

revenue memiliki 128 NaN/Null sedangkan metascore 64. Total ada 192 (catatan tidak overlap)
Maka kurang lebih akan berkurang 192, menyisakan 808 baris data.

### Imputation
Mengganti nilai Null 

```python
#ekstract dulu revenue, agar lebih jelas
revenue = movies_df['revenue_millions']
revenue.head()
```

```python
#ganti dengan mean/ median
revenue_mean = revenue.mean()
revenue_mean
```

```python
revenue.fillna(revenue_mean, inplace=True)
```

<h1 style="font-family: Catamaran Bold; font-size:3em;color:rgb(51, 51, 153);"> Menggunakan Fungsi </p>

```python
# Membuat fungsi 
def rating_function(x):
    if x >= 8.0:
        return "good"
    else:
        return "bad"
```

### Mengaplikasikan fungsi

```python
movies_df["rating_category"] = movies_df["rating"].apply(rating_function)
#menambah kolom baru bernama "rating_category" dengan data rating yang dikenai fungsi

movies_df.head(2)
```

#### Drop kolom

```python
del movies_df['rating_category']
movies_df.head()
```

### Membuat fungsi anonim menggunakan lambda

```python
movies_df["rating_category"] = movies_df["rating"].apply(lambda x: 'good' if x >= 8.0 else 'bad')

movies_df.head(2)
```
